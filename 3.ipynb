# Решение уравнения x'=kx методом Эйлера

```python

import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
# Определяем функцию, задающую параметры ДУ
def dxdt(x,t):
    k=-0.5
    return k*x
t=np.arange(0, 100 , 0.1)
# Решаем ДУ с помощью функции solve
solve_1=odeint(dxdt,y0=1,t=t)
solve_1=solve_1.flatten()
# Определяем функцию, задащую параметры ДУ для решения методом Эйлера
def euler(f, y0=1, h=0.01, n = 100):
    eul = np.arange(0,n,h)
    y = [y0]
    for i in range(len(eul)):
        y.append(y[i] + h*f(eul[i],y[i]))
    return y[:-1], eul
solve_2, eul = euler(dxdt)
# Строим график решения ДУ методом Эйлера
%matplotlib inline
plt.plot(t, solve_1, label=r"$x'=kx$", color = "green")
plt.plot(eul, solve_2, label = r"Метод Эйлера", color = "red")
plt.grid(True)
plt.legend()
plt.show()
```
